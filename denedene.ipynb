```python
import pandas as pd

# Veriyi yükleme
data = pd.read_csv("C:/Users/edana/Desktop/datathon-2024/train.csv")

# Veri tiplerini kontrol etme
print(data.dtypes)

# Sayısal veri sütunları için ortalama veya medyan ile doldurma
# 'Universite Not Ortalamasi' ve 'Lise Mezuniyet Notu' sayısal veri tipinde olmalı
data['Universite Not Ortalamasi'] = pd.to_numeric(data['Universite Not Ortalamasi'], errors='coerce')
data['Lise Mezuniyet Notu'] = pd.to_numeric(data['Lise Mezuniyet Notu'], errors='coerce')

# Ortalama ile doldurma
data['Universite Not Ortalamasi'].fillna(data['Universite Not Ortalamasi'].mean(), inplace=True)

# Medyan ile doldurma
data['Lise Mezuniyet Notu'].fillna(data['Lise Mezuniyet Notu'].median(), inplace=True)

# Kategorik veri sütunları için mod ile doldurma
# 'Universite Turu' ve diğer kategorik sütunlar
categorical_columns = ['Universite Turu', 'Ikametgah Sehri', 'Burs Aliyor mu?', 'Baska Bir Kurumdan Burs Aliyor mu?',
                       'Girisimcilik Kulupleri Tarzi Bir Kulube Uye misiniz?', 'Profesyonel Bir Spor Daliyla Mesgul musunuz?',
                       'Aktif olarak bir STK üyesi misiniz?', 'Stk Projesine Katildiniz Mi?', 'Girisimcilikle Ilgili Deneyiminiz Var Mi?',
                       'Ingilizce Biliyor musunuz?']

for col in categorical_columns:
    data[col] = data[col].astype(str)  # Kategorik verileri stringe dönüştürme
    data[col].fillna(data[col].mode()[0], inplace=True)

# Önceki değerle doldurma (öneri: bu adımı kullanırken dikkatli olun, bazı sütunlar uygun olmayabilir)
data.fillna(method='ffill', inplace=True)

# Sonraki değerle doldurma (öneri: bu adımı kullanırken dikkatli olun, bazı sütunlar uygun olmayabilir)
data.fillna(method='bfill', inplace=True)

# Eksik verileri tekrar kontrol etme
print(data.isnull().sum())

# Veriyi kontrol etme
print(data.head())
```
